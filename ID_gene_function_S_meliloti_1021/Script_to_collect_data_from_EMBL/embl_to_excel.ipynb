```python
import subprocess as sp
import pandas as pd
```

```python
#file = 'AE006469.embl'
#file = 'AL591688.embl'
file = 'AL591985.embl'

filtered_file = file.split('.')[0] + '_filtered.xlsx'
COMMAND = 'awk  AE006469.embl > AE006469_filtered.embl'
args = ["awk", r'$1 == "FT"', file, '>', filtered_file]
p = sp.Popen(args, stdin = sp.PIPE, stdout = sp.PIPE, stderr = sp.PIPE )

f = open(file)
lines = f.readlines()
f.close()
```

```python
df = pd.DataFrame(columns=['gene', 'locus_tag', 'product', 'protein_id', 'EC_number'])
i = -1
for line in lines:
    if 'gene ' in line:
        #print(line)
        i += 1
        df.loc[i] = ['NA', 'NA', 'NA', 'NA', 'NA']
    else:    
        if i >= 0:
            if 'gene=' in line:
                #print(line.split('=')[1])
                df.loc[i][0] = line.split('=')[1].strip('\"').strip('"\n')
            if 'locus_tag=' in line:
                #print(line.split('=')[1])
                df.loc[i][1] = line.split('=')[1].strip('\"').strip('"\n')
            if 'product=' in line:
                #print(line.split('=')[1])
                df.loc[i][2] = line.split('=')[1].strip('\"').strip('"\n')
            if 'protein_id=' in line:
                #print(line.split('=')[1])
                df.loc[i][3] = line.split('=')[1].strip('\"').strip('"\n')
            if 'EC_number=' in line:
                #print(line.split('=')[1])
                df.loc[i][4] = line.split('=')[1].strip('\"').strip('"\n')

#    if 'DECJ' in line: self.DECJ = line.strip().split()[1]
#    if (len(self.RAJ) + len(self.DECJ)) == 0:
#        self.RAJ = self.pname[1:3] + ':' + self.pname[3:5] + ':00.0000'
#        self.DECJ = self.pname[5:8] + ':' + self.pname[8:10] + ':00.0000'

outputname = file.split('.')[0] + '.xlsx'
df.to_excel(outputname) 

df
```
